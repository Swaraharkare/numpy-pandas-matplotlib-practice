```python
import pandas as pd
import numpy as np
```

```python
#series from list

#list
country=['India','Shrilanka','USA','UK','Nepal','Bhutan']
pd.Series(country)
```

```python
#integers
run=[10,56,87,45,67,94]
pd.Series(run)
```

```python
#custom index
marks=[65,90,45,69,87,94,56]
subject=['English','Maths','Biology','Social Science','Science','Chemistry','Physics']

pd.Series(marks,index=subject)
```

```python
#setting name to series
marks=pd.Series(marks,index=subject,name="Marks of swara")
print(marks)
pd.Series(marks,subject,name="Marks of swara")
```

```python
#series from dict
marks={
"English":65,
"Maths":90,
"Biology":45,
"Social Science":69,
"Science":87,
"Chemistry":94,
"Physics":56,   
}
marks_series=pd.Series(marks,name="Marks of swara")
marks_series
```

```python
#Series Attributes

#size
marks_series.size

#dtype---type of series
marks_series.dtype

#name---give name which can we give
print(marks_series.name)

#is_unique----if items unique give true else false
print(marks_series.is_unique)

#index----give index object
print(marks_series.index)

#values---gives values
print(marks_series.values)
```

```python
import pandas as pd
import numpy as np
```

```python
#with one col----aa.txt is one col in dataset-----squeeze is convert dataframe into series when dataset contain one col
cars = pd.read_csv("aa.txt").squeeze("columns")
type(cars)
```

```python
#with two cols
cars = pd.read_csv("sample.txt")
print(cars)
print(type(cars))
```

```python
#with two col----sample.txt have two cols in dataset--stack() is convert dataframe into series when dataset contain multiple cols
df = pd.read_csv("sample.txt",index_col='Car_Model')
# Convert to Series
series = df.stack()
print(type(series))
print(series)
```

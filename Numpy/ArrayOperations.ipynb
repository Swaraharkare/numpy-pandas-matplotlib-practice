```python
import numpy as np
a1=np.arange(12).reshape(3,4)
a2=np.arange(12,24).reshape(3,4)
print(a1)
print(a2)
```

```python
#scalar operations---operation on one numpy array
print(a1)

#arithmetic operation
print(a1*2)
print(a1**2)

#relational
print(a2>15)
print(a2>5)
print(a1==10)

```

```python
#vector---operation two numpy array

#arithmetic
print(a1+a2)
print(a1*a2)

#relational
print(a1<a2)
print(a1>a2)
print(a1+2==a2)
```

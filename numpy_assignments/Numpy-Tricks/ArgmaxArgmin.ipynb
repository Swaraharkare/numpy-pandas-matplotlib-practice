```python
import numpy as np
a=np.array([10,23,45,98,55,34,90,4,67,87,23,76,12,56,99])
```

```python
#returns indices(index) of max element of array in particular axis
np.argmax(a)
```

```python
b=np.random.randint(1,100,24).reshape(6,4)
print(b)
```

```python
#columnwise
np.argmax(b,axis=0)
```

```python
#row wise
np.argmax(b,axis=1)
```

```python
#returns indices(index) of min element of array in particular axis
np.argmin(a)
```

```python
#columnwise
np.argmin(b,axis=0)
```

```python
#row wise
np.argmin(b,axis=1)
```

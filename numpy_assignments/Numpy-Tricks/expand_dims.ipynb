```python
#convert 1d in , 2d in 3d array using expand_dims

import numpy as np

a=np.arange(10)
print(a.shape)

np.expand_dims(a,axis=0)
print(np.expand_dims(a,axis=0).shape)
```

```python
b=np.arange(10,20).reshape(5,2)
print(b.shape)

np.expand_dims(b,axis=0)
print(np.expand_dims(b,axis=0).shape)
```

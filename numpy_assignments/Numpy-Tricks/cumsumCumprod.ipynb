```python
import numpy as np
a=np.array([10,23,45,98,55,34,90,4,67,87,23,76,12,56,99])
b=np.arange(18).reshape(6,3)
b
```

```python
#used when we want to compute cumulative sum of array elements over given axis---it is like sum() window fun in sql
np.cumsum(a)
```

```python
#columnswise
np.cumsum(b,axis=0)
```

```python
#row wise
np.cumsum(b,axis=1)
```

```python
#cumprod
a=np.arange(1,10)

#used when we want to compute cumulative product of array elements over given axis
np.cumprod(a)
```

```python
import numpy as  np

a1=np.random.random((3,3))
a2=np.round(a1*100)
print(a1)
print(a2)
```

```python
#max/min/sum/prod

print(np.max(a1))
print(np.min(a1))
print(np.sum(a1))
print(np.prod(a1))

#to cal max row wise----axis 1 means row and 0 means col
print(np.max(a1,axis=1))
print(np.max(a1,axis=0))
```

```python
#mean/median/std/var

print(np.mean(a1))
print(np.median(a1))
print(np.std(a1))
print(np.var(a1))



#by colwise or row wise
print(np.mean(a1,axis=1))
print(np.median(a1,axis=1))
print(np.std(a1,axis=1))
print(np.var(a1,axis=1))
```

```python
#trignometric function

print(np.sin(a1))
```

```python
#dot product
a1=np.arange(12).reshape(3,4)
a2=np.arange(12,24).reshape(4,3)

print(np.dot(a1,a2))
```

```python
#log and exponents

print(np.log(a2))
print(np.exp(a1))
```

```python
#round/floor/ceil
a1=np.random.random((2,3))*100
print(a1)
print(np.round(a1))
print(np.floor(a1))
print(np.ceil(a1))
```

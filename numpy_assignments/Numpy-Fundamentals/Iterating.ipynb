```python
import numpy as np
a1=np.arange(10)
a2=np.arange(12).reshape(3,4)

print(a1)
print(a2)
```

```python
for i in a1:
    print(i)
```

```python
for i in a2:
    print(i)
```

```python
for i in a2:
    for j in i:
        print(j)
```

```python
# using nditer()

for i in np.nditer(a2):
    print(i)
```

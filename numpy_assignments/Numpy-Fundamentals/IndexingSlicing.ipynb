```python
import numpy as np

a1=np.arange(10)
a2=np.arange(12).reshape(3,4)
a3=np.arange(8).reshape(2,2,2)
a2
```

```python
##indexing
```

```python
#1D
print(a1)
print(a1[1])
print(a1[-2])
```

```python
#2D
print(a2)
print(a2[2,2])
print(a2[1,2])
```

```python
#3D
print(a3)

print(a3[1,0,1])
print(a3[0,0,1])
print(a3[0,0,0])
print(a3[1,1,0])
```

```python
##slicing
```

```python
#1D
print(a1)
print(a1[2:5])
```

```python
##2D
```

```python
#2D
print(a2)

#to fetch first row
print(a2[0,:])

#to fetch 3rd col
print(a2[:,2])

#to print 5 6 9 10
print(a2[1:,1:3])

#to print elements from corner 0 3 8 11
print(a2[::2,::3])

#to print 1 3 9 11
print(a2[::2,1::2])

##to print 4 7
print(a2[1,::3])
```

```python
##3D
```

```python
#3D
a3=np.arange(27).reshape(3,3,3)
print(a3)

#fetch center 2D array
print(a3[1])

#
print(a3[::2])
```

```python
print(a3)

#fetch 3 4 5
print(a3[0,1,:])
```

```python
#fetch 10 13 16
print(a3[1,:,1])
```

```python
#fetch 22 23 25 26
print(a3[2,1:,1:])
```

```python
#fetch 0 2 18 20
print(a3[::2,0,])
```

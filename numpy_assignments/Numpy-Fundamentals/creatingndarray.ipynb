```python
#np.array
import numpy as np
a=np.array([1,2,3])
print(a)
print(type(a))
```

```python
#2D and 3D
b=np.array([[1,2,3],[3,4,5]])
print(b)
c=np.array([[1,2,3],[3,4,5],[4,5,6]])
print(c)
```

```python
#dtype
a=np.array([1,2,3], dtype=float)
print(a)

a=np.array([1,2,3], dtype=bool)
print(a)

a=np.array([1,2,3], dtype=complex)
print(a)
```

```python
#np.arange
np.arange(10)
```

```python
#with reshape()
np.arange(1,11).reshape(5,2)
```

```python
#np.ones and np.zeros
ones=np.ones((3,4))
print(ones)

zeros=np.zeros((3,4))
print(zeros)
```

```python
#np.random()

np.random.random((3,4))
```

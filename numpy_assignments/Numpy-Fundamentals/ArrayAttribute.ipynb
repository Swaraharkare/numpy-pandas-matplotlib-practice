```python
import numpy as np
a1=np.arange(10)
a2=np.arange(12,dtype='float').reshape(3,4)
a3=np.arange(8).reshape(2,2,2)
a3
```

```python
#.ndim

print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
```

```python
#shape

print(a1.shape)
print(a2.shape)
print(a3.shape)
```

```python
#size---how many elements in array

print(a1.size)
print(a2.size)
print(a3.size)
```

```python
#itemsize--

print(a1.itemsize)
print(a2.itemsize)
print(a3.itemsize)
```

```python
#dtype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

```python
#change data type
#astype

print(a3.dtype)
a3.astype(np.int32)
```

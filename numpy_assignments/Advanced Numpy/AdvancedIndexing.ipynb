```python
#Advanced Indexing

import numpy as np
a=np.arange(24).reshape(6,4)
print(a)
```

```python
#fancy indexing

#to fetch rows
print(a[[0,2,3,5]])
```

```python
#to fetch cols
print(a[0:,[0,2,3]])
```

```python
#Boolean Indexing
a2=np.random.randint(1,100,24).reshape(6,4)
print(a2)
```

```python
#find all numbers from a2 which are greater than 50
print(a2[a2>50])
```

```python
#find out even numbers
print(a2[a2%2==0])
```

```python
#find all numbers greater than 50 and are even
print(a2[(a2>50) & (a2%2==0)])

#find all numbers greater than 50 or are even
print(a2[(a2>50) | (a2%2==0)])
```

```python
#find all numbers not divisible by 7
print(a2[a2%7 !=0])
print(a2[~(a2%7==0)])

#find all numbers divisible by 7
print(a2[a2%7==0])
```

```python
#Broadcasting
import numpy as np
a=np.arange(6).reshape(2,3)
print(a)
b=np.arange(6,12).reshape(2,3)
print(b)
print(a+b)
```

```python
#diff shape
a=np.arange(6).reshape(2,3)
print(a)
b=np.arange(3).reshape(1,3)
print(b)
print(a+b)
```
